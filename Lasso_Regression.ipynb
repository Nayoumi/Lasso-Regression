```python
lasso_features=lasso_dataset.drop(['superplasticizer','csMPa'],axis=1)
lasso_target=lasso_dataset['csMPa']
# splitting the dataset into train and test
```

**Standardization:**
Its is used when the data is inthe form of Guassian or normal distribution. But not necessarily to be in normal distribution. It is used when we want zero mean and standard deviation is unit. Mean and standard devation is used to scale the values. It is also called as z-score normalization.

*it standardizes by removing the mean or mean is centered and scaling to unit standard deviation. less sensitive to outliers.
   
                   standardization= x-mean/standard deviation

Implementing accuracy for training and testing data for different test sizes and alpha values

```python
lasso_train_score=[]
lasso_test_score=[]
for i in range(1,5):
  size=i/10
  scaler= StandardScaler() # standarizing the model
  #lasso_features=scaler.fit_transform(lasso_features)
  X_train,X_test,y_train,y_test=train_test_split(lasso_features,lasso_target,test_size=size) #splitting the data set ro train and test
  lasso_model=Lasso(alpha=size) # Building the model with alpha values 0.1 to 05
  pipe=make_pipeline(scaler,lasso_model)
  pipe.fit(X_train,y_train) # tarning the model
  train_score=pipe.score(X_train,y_train)# Accuarcy of the training data
  test_score=pipe.score(X_test,y_test)# Accuracy of the testing data
  lasso_train_score.append(round((train_score*100),0))
  lasso_test_score.append(round((test_score*100),0))
print("Accuracy of training set:",lasso_train_score)
print("Accuracy of testing set: ",lasso_test_score)

#for test_size of 0.2 we can observe that training and testing score as same
```

```python
lasso_train_score=[]
lasso_test_score=[]
for i in range(1,5):
  size=i/10
  scaler= StandardScaler() # standarizing the model
  #lasso_features=scaler.fit_transform(lasso_features)
  X_train,X_test,y_train,y_test=train_test_split(lasso_features,lasso_target,test_size=size) #splitting the data set ro train and test
  lasso_model=Lasso(alpha=size/10) # Building the model with alpha values 0.10 to 0.05
  pipe=make_pipeline(scaler,lasso_model)
  pipe.fit(X_train,y_train) # tarning the model
  train_score=pipe.score(X_train,y_train)# Accuarcy of the training data
  test_score=pipe.score(X_test,y_test)# Accuracy of the testing data
  lasso_train_score.append(round((train_score*100),0))
  lasso_test_score.append(round((test_score*100),0))
print("Accuracy of training set:",lasso_train_score)
print("Accuracy of testing set: ",lasso_test_score)

#for test_size of 0.3 we can observe that training and testing score as same
```

Implemeting accuarcy and MSE for different alpha values

```python
lasso_score=[]
lasso_mse=[]
for i in range(1,11):
  lasso_alpha=i/10
  lasso_model=Lasso(alpha=lasso_alpha)
  lasso_model.fit(X_train,y_train)
  accuracy =lasso_model.score(X_test,y_test)
  lasso_score.append(round((accuracy*100),2))
  lasso_predict=lasso_model.predict(X_test)
  mse=mean_squared_error(lasso_predict,y_test)
  lasso_mse.append(round(mse,2))
print('\033[1m'+"Accuracy and MSE for different alpha values from 0.1 to 1"+'\033[0m')
print("Accuracy:",lasso_score)
print("MSE: ",lasso_mse)
```
